# NFL 2023 Survivor Pool Tool

## Get the Schedule:
from website: https://dknetwork.draftkings.com/nfl/2023/5/12/23680526/nfl-schedule-release-printable-full-game-list-google-doc-excel-every-game-matchup-print-out

```python
import pandas as pd

sched = pd.read_excel("2023_NFL_Schedule_Grid.xlsx", index_col=0)
sched
```

```python
sched.columns
```

```python
sched.columns = ['Wk01', 'Wk02',
       'Wk03', 'Wk04', 'Wk05',
       'Wk06', 'Wk07', 'Wk08',
       'Wk09', 'Wk10', 'Wk11',
       'Wk12', 'Wk13', 'Wk14',
       'Wk15', 'Wk16', 'Wk17', 'Wk18']
sched['Wk01'].loc[sched.Wk01.str.contains('vs. ')]
```

```python
sched.index.values
```

```python
sched.index = ['Cardinals', 'Falcons', 'Ravens',
       'Bills', 'Panthers', 'Bears',
       'Bengals', 'Browns', 'Cowboys',
       'Broncos', 'Lions', 'Packers',
       'Texans', 'Colts', 'Jaguars',
       'Chiefs', 'Raiders', 'Chargers',
       'Rams', 'Dolphins', 'Vikings',
       'Patriots', 'Saints', 'Giants',
       'Jets', 'Eagles', 'Steelers',
       '49ers', 'Seahawks', 'Buccaneers',
       'Titans', 'Commanders']
sched.index.name = 'Home_Team'
sched
```

```python
for i in sched.columns: 
    sched[i].loc[sched[i].str.contains('@ ')]=''
    sched[i].loc[sched[i].str.contains('BYE')]=''
    sched[i] = sched[i].str.replace('vs. ', '', regex=False)
    sched[i] = sched[i].str.replace('vs ', '', regex=False)
    sched[i] = sched[i].str.replace(' (MNF)', '', regex=False)
    sched[i] = sched[i].str.replace(' (SNF)', '', regex=False)
    sched[i] = sched[i].str.replace(' (TNF)', '', regex=False)
    sched[i] = sched[i].str.replace(' (Thanksgiving)', '', regex=False)
    sched[i] = sched[i].str.replace(' (London)', '', regex=False)
    sched[i] = sched[i].str.replace(' (Germany)', '', regex=False)
sched
```

```python
#Some Sanity Checks:
g = 0
for i in sched.index.values:
    n = 0
    for j in sched.columns:
        l = sched[j].str.contains(i).sum()
        n = n + l
        g = g + l
    print(i, n, 'road games')

print(g, 'total games counted')
print('there are 272 games in the 2023 NFL Season')
```

```python
recs = pd.read_html("https://www.nfl.com/standings/league/2022/reg", header=0, index_col=0)
records = pd.DataFrame(recs[0])
records.sort_values(by="Net Pts", ascending=False)
```
